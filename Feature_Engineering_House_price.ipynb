```python
#let's see one machisim handling missing values

## Replacing categorical features
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
categorical_features
```

```python
#1st technique
data=df.copy()
def replace_cat_features(dataset, categorical_features):
    data[categorical_features]=data[categorical_features].fillna('Missing')
    return data
    
```

# Numerical Nan values

```python
[feature for feature in df.columns if df[feature].dtype!='O' and df[feature].isnull().sum()>=1]
```

```python
numerical_features_nan=[feature for feature in df.columns if df[feature].dtype!='O' and df[feature].isnull().sum()>=1]
numerical_features_nan
```

```python
for feature in numerical_features_nan:
    ## since there are outliers we are going to replace with median
    median_values=df[feature].median()
    ## create a new feature to capture nan value
    df[feature+'_nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_values,inplace=True)
```

```python
df.head()
```

```python
df['LotFrontage'].isnull().sum()
```

```python
## temporal Variables( date time variable)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[feature+'_age of house']=df['YrSold']-df[feature]
```
