There are so many outlier in the dependent feature. That's why we are using median()

```python
# How to replace the missing values in Feature Engineering and what is the relationship between Indepdent and dependent varibales
## let's plot some amazing diagram
data=df.copy()

for feature in features_with_na:
    # let's make a variables that indicate 1 if the observation was missing or 0 otherwise
    data[feature]=np.where(data[feature].isnull(),1,0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
    
```

Here with the relationship between the `missing values and the dependent variable is clearly visible`. So, we need to replace these `nan` values with something meaningful which we will do in the `Feature Engineering`.

```python
## list of the numerical variables
[feature for feature in df.columns if df[feature].dtype!='o']#show only numerical features
```

```python
numerical_features=[feature for feature in df.columns if df[feature].dtype!='O']#show only numerical features
```

```python
print(len(numerical_features))
df[numerical_features].head()
```

```python
df.info()
```

```python
## Temporal Variables(Eg: Datetime Variables)
# it can be int64
# Relation between Datetime and SalePrice, this is very important.
[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
year_feature=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
for feature in year_feature:
    print(feature,df[feature].unique())#only show the unique value
```

```python
# this is very very important. relationship between time and Saleprice
# price can be increased or decreased based on time 

df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
plt.show()
```

```python
### here we will compare the difference between all years features with SalePrice
data=df.copy()
for feature in year_feature:
    if feature!='YrSold':
        data[feature]=data['YrSold']-data[feature]#this is age of the house
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

### Observations
Within all the yearwise information, it is clearly visible that the new houses are more costlier that the older one. looks like a pareto or power law distribution.

```python
### Numerical Variables
##1. Continous Variables: This can be anything. 
##2. Discreate Variable: this is whole number;it can be limited like <=25

[feature for feature in numerical_features if len(df[feature].unique())<=25]
```

```python
discrete_feature=[feature for feature in numerical_features if len(df[feature].unique())<=25]
print(len(discrete_feature))
df[discrete_feature].head()
```

```python
## let's find the relationship between Discrete and Sales Price
data=df.copy()
for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sales Price')
    plt.title(feature)
    plt.show()
```

Observations:

There is a relationship between Discreate features and Sale Price

```python
## Continous Variable

[feature for feature in numerical_features if feature not in discrete_feature+year_feature]
```

```python
continous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature]
print(len(continous_feature))
df[continous_feature].head()
```

```python
## let's analyze the continous values by creating histograms to understand the distribution
data=df.copy()
for feature in continous_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel('Sales Price')
    plt.title(feature)
    plt.show()
    
```

```python
# best way for continous variable as hist plot
data=df.copy()
for feature in continous_feature:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```
